# Double peak triangular distribution

Let's plot a triangular distribution first.

```python
import numpy as np
import matplotlib.pyplot as plt

n = 100

def triangularFunction(xval, a):
    
    # Double triangle, each centered in -a and +a
    # Peaks at 1/a
    
    if xval < -a:
        return 0
    elif xval >= -a and xval < 0:
        return  1/(a)**2 * xval + 1/a
    elif xval >= 0.    and xval < a:
        return -1/(a)**2 * xval + 1/a
    elif xval >= a:
        return 0.
    else:
        return -1.
  
x = np.linspace(-5.,+5., n)

aVals = [1, 2, 3, 4, 5]
colors= ['m','brown','r','orange','gold']

plt.figure(figsize=(10,5))
plt.xlabel("x values")
plt.ylabel("Density")

for aVal in aVals:
    y = [ triangularFunction(xi, aVal) for xi in x]
    plt.plot(x,y,color=colors[aVal-1], label='a = %d'%aVal)
plt.title('Triangular distributions for several limits a')
plt.legend()
plt.show()

```

Now double peaks!

```python
def doubleTriangularFunction(xval, a):
    
    # Double triangle, each centered in -a/2 and +a/2, from -a ; a total spread.
    # Peaks at 1/a
    if a < 0:
        return -99
    
    if xval < -a:
        return 0
    
    elif xval >= -a  and xval < -a/2.:
        return 2/(a)**2 * xval + 2/a
    
    elif xval >= -a/2. and xval < 0:
        return -2/(a)**2 * xval 
    
    elif xval >= 0.  and xval < a/2.:
        return 2/(a)**2 * xval 
    
    elif xval >= a/2. and xval < a:
        return -2/(a)**2 * xval + 2/a
    
    elif xval >= a:
        return 0.
    
    else:
        return -1.

n = 1000
x = np.linspace(-5.,+5., n)

plt.figure(figsize=(10,5))
plt.xlabel("x")
plt.ylabel("y")

for aVal in aVals:
    y = [ doubleTriangularFunction(xi, aVal) for xi in x]
    plt.plot(x,y,color=colors[aVal-1], label='a = %d'%aVal)
plt.title('Double triangular distribution of limit a')
plt.legend()
plt.show() 
    
```
